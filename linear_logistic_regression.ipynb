### Cleaning and loading dataset

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,recall_score,f1_score,precision_score,confusion_matrix,classification_report
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
titanic_data = pd.read_csv(r'tested.csv')
titanic_data.head()
```

```python
titanic_data.info()
```

```python
titanic_data.describe()
```

```python
titanic_data['Age']= titanic_data['Age'].fillna(titanic_data['Age'].median())
titanic_data.info()
```

```python
titanic_data['Fare']= titanic_data['Fare'].fillna(titanic_data['Fare'].median())
titanic_data.info()
```

```python
titanic_data = titanic_data.drop(columns='Cabin', axis=1)
titanic_data.head()
```

```python
identity = titanic_data[['PassengerId','Name','Ticket']]
identity.head()
```

### Onehotencoding get dummies pclass and embarked 
- nominal data 

```python
titanic_data = pd.get_dummies(titanic_data,columns=['Pclass','Embarked'], drop_first=True, dtype=int)
titanic_data.head()
# we drop to avoid multicollinearity
```

### Label encoding sex column
- ordinal and binary data

```python
le = LabelEncoder()
titanic_data['Sex']= le.fit_transform(titanic_data['Sex'])
titanic_data.head()
```

```python
gender_survived = titanic_data.groupby('Sex')['Survived'].agg('count')
gender_survived.head()
```

```python
survived= titanic_data[titanic_data['Survived']==1]
survived.head()
```

```python
gender_survivors = survived.groupby('Sex')['Survived'].agg('count')
gender_survivors.head()
```

```python
survived['Sex'].value_counts()
```

```python
y = titanic_data['Survived']
x = titanic_data.drop(['Survived','PassengerId','Name','Ticket'],axis=1)
x.head()
```

### Splitting our data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=22)
```

- Apply scaling and class weight balancing in the y column and evaluate the model performance
- Also fit logit (by statsmodels) and compare model performance
- class weight - imblearn -imbalanced learn(SMOTE)

### Fitting the data using logistic regression

```python
lr = LogisticRegression()
model = lr.fit(x_train,y_train)
```

```python
x.info()
```

### Metrics

```python
y_pred = model.predict(x_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
precision_score(y_pred,y_test)
```

```python
recall_score(y_pred,y_test)
```
